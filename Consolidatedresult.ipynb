```python
import pandas as pd
```

```python
# Upload and read each strategy/model result
rsi = pd.read_csv("rsi_results.csv")
rsi['Model'] = 'RSI'

macd = pd.read_csv("macd_results.csv")
macd['Model'] = 'MACD'

boll = pd.read_csv("BollingerBands_Performance.csv")
boll['Model'] = 'Bollinger Bands'

xgb = pd.read_csv("XGBoost_Stock_Performance.csv")
xgb['Model'] = 'XGBoost'

lstm = pd.read_csv("lstm_results.csv")
lstm['Model'] = 'LSTM'

tcn = pd.read_csv("tcn_results.csv")
tcn['Model'] = 'TCN'

trans = pd.read_csv("transformer_results.csv")
trans['Model'] = 'Transformer'

# Combine all
final_df = pd.concat([rsi, macd, boll, xgb, lstm, tcn, trans])
final_df = final_df[['Model', 'Stock', 'Accuracy', 'Sharpe Ratio', 'Max Drawdown (%)']]

# Save and download
final_df.to_csv("Final_Combined_Results.csv", index=False)

from google.colab import files
files.download("Final_Combined_Results.csv")

# Display
final_df
```
